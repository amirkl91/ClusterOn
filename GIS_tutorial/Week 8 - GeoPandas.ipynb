# GeoPandas - It's just pandas with an addiotinal column of shapely geometric objects!!

## Import packages - not included in anaconda.

```python
import geopandas as gpd
%matplotlib inline
```

### import ready-made example geoDateFrame of the world

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

#### verify the type - GeoDataFrame indeed!

```python
type(world)
```

#### explore the GDF head

```python
world.head(20)
```

#### check how many entries in this table

```python
len(world)
```

```python
world.crs
```

#### plot table geometries. as simple as that

```python
world.plot(figsize = (8,10))
```

```python
world = world.to_crs("EPSG:3395")
```

```python
world.crs
```

```python
world.plot(figsize = (8,10))
```

```python
world = world[(world.name != "Antarctica") & (world.name != "Fr. S. Antarctic Lands")]
```

```python
world.plot(figsize = (8,10))
```

```python
world = world.to_crs("EPSG:4326")
```

```python
world.crs
```

```python
world.columns
```

#### Plot choropleth of the world's Median GDP by countries

```python
world.plot(column = 'gdp_md_est',
          figsize = (8,10))
```

#### add legend

```python
world.plot(column = 'gdp_md_est',
          figsize = (8,10),
          legend = True)
```

#### color by quantiles

```python
world.plot(column = 'gdp_md_est',
          figsize = (8,10),
        scheme = 'Quantiles',
          legend = True)
```

#### make it 10 quantiles

```python
world.plot(column = 'gdp_md_est',
          figsize = (8,10),
           k = 10,
        scheme = 'Quantiles',
          legend = True)
```

#### filter geoDataFrame for one continent

```python
world[world['continent']=='Africa'].plot(column = 'gdp_md_est',
                                        figsize = (8,10),
                                        k = 10,
                                        scheme = 'Quantiles',
                                        legend = True)
```

## Plotting one map using mutliple geoDataFrames

### import cities in the world

```python
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
```

```python
cities.head()
```

### import pyplot for plotting...

```python
from matplotlib import pyplot as plt
```

#### Plot

```python
# a shortcut to create a figure and an axes object in one line
fig, ax = plt.subplots(figsize=(15,10))

# make axes aspect equal
ax.set_aspect('equal')

# Plot world countries, without Antarctica
world[world['continent']!='Antarctica'].plot(ax = ax,             # use axes ax                                      
                                             color = 'white',     # paint all with white
                                             edgecolor ='black')  # edges black please

# Plot cities
cities.plot(ax = ax,      # use the same axes - ax
            color = 'blue', # pick color blue
            marker = '*',   # marker style - *
            markersize=40)  # marker size 40
# add grid
plt.grid()
# fix layout
plt.tight_layout()
```

## Spatial join - mergind two geoDataFrame based on a geometry field

```python
cities.head()
```

```python
world.head()
```

#### join world with cities table, where the cities intersect with the world polygons.

```python
result = gpd.sjoin(cities, world)
```
